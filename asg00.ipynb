```python
# Data types do not require a specific type

myString = "Hello World"
myInt = 5
myFloat = 5.5
myList = [1,2,3,4,5]
myTuple = (myString,myInt)
myDict = {
    "Auto" : 100,
    "Haus" : 200,
}

print("myString = ", myString)
print("myInt = ", myInt)
print("myFloat = ", myFloat)
print("myList = ", myList)
print("myDict = ", myDict)
print("myTuple = ", myTuple)
```

```python
# If statements
foo = 5
bar = 10

# Note the tabs for intendation!
if foo <= bar:
    print("foo <= bar")
else:
    print("foo > bar")
    
mybool = foo <= bar

# True and False are written with a captial letter
if mybool is True: # use '==' as an alternative
    print("mybool is True")
elif foo <= 100:
    print("WDWDWD")
else:
    print("mybool is False")
    
# For loops
for i in range(1,10):# similar to Javas: for (int i = 0; i < 10; ++i)
    print("i = ", i)
print()

# For loops with iterators
for e in myList: 
    print("e = ", e)

print()

# For loops with iterators
for i in range(0,len(myList)): 
    print("myList[", i, "] =", myList[i])
    
print()

# While loops
i = 0
while i < 10:
    print("i = ", i)
    i += 1
```

```python
# function definitions. Ne specific return type required - just return the value

def inner(x1,x2):
    tmp = 0
    for i,j in zip(x1,x2):
        tmp += i*j
    return tmp

print("<[1,2],[3,4]> = ", inner([1,2],[3,4]))
print("<[-5,3],[9,4]> = ", inner([-5,3],[9,4]))
```
